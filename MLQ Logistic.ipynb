```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Plotting Library
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

data =  pd.read_csv(r'C:\Users\dell\ML\SMSSpamCollection.csv', sep='\t', header=None, names=['Label', 'SMS'])
data
```

```python
(X_train_raw, X_test_raw, y_train, y_test) = train_test_split(data['SMS'], data['Label'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train_raw)
X_test = vectorizer.transform(X_test_raw)

testing_data = pd.DataFrame()
testing_data['SMS'] = X_test_raw
testing_data['Label'] = y_test
```

```python
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
predictions = classifier.predict(X_test)
testing_data['predicted'] = predictions
testing_data
```

```python
total = len(testing_data)
correct = 0
```

```python
for row in testing_data.iterrows():
    actual_classifcation = row[1]['Label']
    predicted_classifcation = row[1]['predicted']
    if actual_classifcation == predicted_classifcation:
        correct += 1
```

```python
accuracy = correct / total
print(accuracy)
```
