```python
import numpy as np
import pandas as pd
from sklearn import linear_model

import math

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (8,6)

import sys
sys.path.append(r'C:\Users\Tina_\Desktop\Machine learning\ml-networking-master\lib') 
from parse_pcap import pcap_to_pandas, send_rates

pcap = pcap_to_pandas(r'C:\Users\Tina_\Desktop\Machine learning\Data\test2.pcap')
```

```python
pdf = pcap.loc[:,['ip_src','ip_dst','length','port_src','port_dst','protocol']]
pdf.head(20)
```

```python
pdf1 = pdf.dropna()
pdf1.head(20)
```

```python
pdf_bytes = pdf1.groupby(['ip_src','ip_dst','port_src','port_dst','protocol']).sum()
pdf_pkts = pdf1.groupby(['ip_src','ip_dst','port_src','port_dst','protocol']).count()
```

```python
pkts = pdf_pkts['length'].values
byte = pdf_bytes['length'].values

points_x = pkts
points_y = byte
```

```python
###########
# Step 1: Create Linear Regression Model
linear_regression = linear_model.LinearRegression()

# Step 2: Train the Model Using the (x,y) Points
# Note: Reshape (-1,1) creates a column vector, which is what fit() expects.
linear_regression.fit(points_x.reshape(-1,1), points_y.reshape(-1,1))

# Step 3: Input to predict function are x values that the model has never seen.
y_hat = linear_regression.predict(np.array([-2,30]).reshape(-1,1))
###########

# Plot the results along with the points and previous line
# Plot the original points.
plt.plot(points_x, points_y, '.', color='blue', markersize=12)

# Predicted line from sklearn is in red
plt.plot([-2, 30], y_hat, color='red', alpha=0.4, linewidth=3)

plt.grid(linestyle='--', alpha=0.6)
plt.xlim((-1, 30))
plt.xlabel('Packets')
plt.ylabel('Bytes')
plt.show()
```

```python
pkts = 9
byte = linear_regression.predict([[pkts]])
print("Predicted byte with {0} pkts : {1:.2f} ".format(pkts, byte[0][0]))
```

```python
distances = np.sqrt((points_y - y_hat)**2)   # distances from y values of the points to y values of the line
total_distance = np.sum(distances)            # sum of the distances
print("Residual Sum of Squares (RSS): {:.4f}".format(total_distance))
```
